# Partie 1
Le but de cette partie est de générer un fichier de data pour pouvoir entrainer notre modèle. !!! Peu de données pour le moment

```python
import os
import cv2
import numpy as np

# Chemin du dossier contenant les images
dossier_images = "image_source/caractere/"

# Assurez-vous que le dossier de sortie existe, sinon créez-le
if not os.path.exists(dossier_images):
    os.makedirs(dossier_images)

# Liste des extensions d'images que vous souhaitez traiter
extensions_images = [".jpg", ".jpeg", ".png"]

# Initialiser une liste pour stocker les tableaux NumPy des images
tableaux_images = []
noms_fichiers_sans_extension = []  # Nouvelle liste pour stocker les noms de fichiers sans extension

# Boucle sur les fichiers dans le dossier d'images
for fichier in os.listdir(dossier_images):
    if fichier.endswith(tuple(extensions_images)):
        chemin_image = os.path.join(dossier_images, fichier)

        # Charger l'image avec OpenCV
        image = cv2.imread(chemin_image)

        # Assurez-vous que l'image est lue avec succès
        if image is not None:
            # Convertir l'image en tableau NumPy
            image_array = np.array(image)

            # Ajouter le tableau NumPy à la liste
            tableaux_images.append(image_array)
            
            # Ajouter le nom du fichier sans extension à la liste
            nom_fichier_sans_extension = os.path.splitext(fichier)[0]
            noms_fichiers_sans_extension.append(nom_fichier_sans_extension)

            print(f"L'image {chemin_image} a été ajoutée à la liste.")
        else:
            print(f"Impossible de charger l'image à partir de {chemin_image}")

# Empiler les tableaux NumPy le long du nouvel axe (3e axe)
tableau_final = np.stack(tableaux_images, axis=0)

# Enregistrer le tableau NumPy dans un fichier .npy
fichier_npy_final = "./images.npy"
np.save(fichier_npy_final, tableau_final)

# Enregistrer les noms des fichiers sans extension dans un fichier .npy
fichier_noms_fichiers = "./noms_fichiers.npy"
np.save(fichier_noms_fichiers, np.array(noms_fichiers_sans_extension))

print(f"Le fichier .npy final a été créé avec succès : {fichier_npy_final}")
print(f"Le fichier .npy des noms de fichiers a été créé avec succès : {fichier_noms_fichiers}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

# Charger les données à partir des fichiers .npy
fichier_npy_images = new_X
fichier_npy_noms_fichiers = new_Y

X = new_X
noms_fichiers = new_Y

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(noms_fichiers)

# Aplatir chaque image
X_flat = X.reshape(X.shape[0], -1)

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_flat, y, test_size=0.2, random_state=42)

# Construire le pipeline avec PCA et SVM
pipeline = Pipeline([
    ('pca', PCA(n_components=50)),  # Réduire les dimensions avec PCA (ajuster le nombre de composantes selon les besoins)
    ('svm', SVC(kernel='linear', C=1))  # Utiliser un SVM linéaire (vous pouvez ajuster le noyau et les hyperparamètres)
])

# Entraîner le modèle
pipeline.fit(X_train, y_train)

# Prédire sur l'ensemble de test
y_pred = pipeline.predict(X_test)

# Calculer la précision
precision = accuracy_score(y_test, y_pred)
print(f"Précision du modèle : {precision}")
```

```python
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder

model = pipeline.named_steps['svm']
label_encoder = label_encoder  # Remplacez par le nom réel de votre LabelEncoder

# Charger une nouvelle image pour la prédiction
nouvelle_image_path = "/home/baptistedie/Etud/ProjetPapi/image_source/caractere/C1.jpg"
nouvelle_image = cv2.imread(nouvelle_image_path)

# Prétraiter l'image de la même manière que les données d'entraînement
nouvelle_image = nouvelle_image.reshape((1, -1))  # Aplatir l'image
nouvelle_image_pca = pipeline.named_steps['pca'].transform(nouvelle_image)  # Appliquer PCA

# Faire la prédiction avec le modèle SVM
prediction = model.predict(nouvelle_image_pca)

# Inverser la transformation du LabelEncoder pour obtenir le nom de la classe prédite
nom_classe_predite = label_encoder.inverse_transform(prediction)

print(f"La nouvelle image est prédite comme étant de la classe : {nom_classe_predite[0]}")
```
